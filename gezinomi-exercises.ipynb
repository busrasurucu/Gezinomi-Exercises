## Convert the 'SaleCheckInDayDiff' variable into a categorical variable.

```python
bins_=[0, 7, 30, 90, df["SaleCheckInDayDiff"].max()]
labels_=["Last Minuters", "Potential Planners", "PLanners", "Early Bookers"]
```

```python
df["check"] = pd.cut(df["SaleCheckInDayDiff"], bins = bins_, labels = labels_, include_lowest = True)
df.head()
```

## What are the average earnings in the breakdown of COUNTRY, SOURCE, SEX, and AGE?

```python
df.groupby(["SaleCityName", "ConceptName", "CheckInDate"]).agg({"Price": ["mean", "count"]})
```

```python
agg_df = df.groupby(["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values(by="Price", ascending=False)
agg_df.head()
```

## Convert the names in the index to variable names.

```python
agg_df = agg_df.reset_index()
agg_df.head()
```

## Define new_level_based customer personas.

```python
agg_df["customer_level_based"] = [f"{s.upper()}_{c.upper()}_{n.upper()}" for s,c,n in zip(agg_df["SaleCityName"], agg_df["ConceptName"],agg_df["Seasons"])]
agg_df.head()
```

## Please solve the same question using a lambda.

```python
agg_df["customer_level_based"] = agg_df[["SaleCityName", "ConceptName", "Seasons"]].agg(lambda x: '_'.join(x).upper(), axis=1)
agg_df.head()
```

## Please solve the same question using f-string and interrows.

```python
agg_df['customer_level_based'] = [f"{row['SaleCityName'].upper()}_{row['ConceptName'].upper()}_{row['Seasons'].upper()}" for index, row in agg_df.iterrows()]
agg_df.head()
```

## Segment new customers (personas).

```python
labels_ = ["D", "C", "B", "A"]
agg_df["Segment"] = pd.qcut(agg_df["Price"], 4, labels= labels_)
agg_df.head()
```
